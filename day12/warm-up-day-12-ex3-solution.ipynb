```python
%matplotlib inline

import numpy as np
import pandas as pd
import numpy as np
import matplotlib as mpl

# create the xaxis points
xaxis = [np.random.randint(-1,1)/10 + np.cos(t)**3 for t in np.arange(0,8, .05)]

# create the yaxis points
yaxis = [np.random.randint(-1,1)/10 + np.sin(t)**3 for t in np.arange(0,8, .05)]

# given a Normalize object, which will map to valid colors in a color map given values 
# spanning from vmin to vmax, create a color map
norm = mpl.colors.Normalize(vmin=0, vmax=8)
cm = mpl.cm.ScalarMappable(norm=norm, cmap='Accent')

# call the color map for the range of the graph, 0 to 8
# if the data frame had another column representing, for instance, intensity of plot point,
# this could be used to convert that data element to a color representing intensity
colors = [cm.to_rgba(t)  for t in np.arange(0,8, .05)]

# add the colors to the dataframe so if dataframe manipulation occurs the colors will still
# align to the rows they belong
df = pd.DataFrame({'x':xaxis, 'y':yaxis, 'color':colors})

# use the data frame color column for the color of points
plot = df.plot(kind='scatter', x = 'x', y = 'y', c = df['color'] )

plot.set_xlabel('')
plot.set_ylabel('')

plt.show()
```
