```python
%matplotlib inline

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

xaxis = [(t**2)/(np.e**t) for t in np.arange(0,8, .15)]

yaxis = [t / (np.e**t) for t in np.arange(0,8, .15)]

df = pd.DataFrame({'x':xaxis, 'y':yaxis})

df = df.set_index('x')

plot = df.plot()

plot.get_legend().remove()

plot.set_xlabel('')

plt.show()
```
