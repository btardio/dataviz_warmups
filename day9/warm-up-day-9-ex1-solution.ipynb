```python
import pandas as pd

alldicts = []

def rgbsplit(colorstring):
    return {'red': colorstring[0] + colorstring[1],
            'green': colorstring[2] + colorstring[3],
            'blue': colorstring[4] + colorstring[5]}
    
colors = ['020000', '010001', '010201', '010200', '020100', '010202', '020101', '020102','020000']

for color in colors:
    
    alldicts.append(rgbsplit(color))
    
df = pd.DataFrame(alldicts)

df[['red','green','blue']]
```

```python
redblue = df.groupby(['red', 'blue']).count()
redblue
```

```python
redgreen = df.groupby(['red','green']).count()
redgreen
```

```python
redblue[redblue['green']>2]
```

```python
redgreen[redgreen['blue']>2]
```
