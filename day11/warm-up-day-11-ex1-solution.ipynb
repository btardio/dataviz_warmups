```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

minx = -2.0
maxx = 2.0
step = 0.05

def myfloatrange(start, stop, mfrstep):
    i = 0
    while ( start + ( i * mfrstep ) < stop ):
        yield round(start + ( i * mfrstep ),2)
        i += 1

yseriesa = []
yseriesb = []
yseriesc = []

for x in myfloatrange(minx, maxx+step, step):
    yseriesa.append( (0 if x == 0 else np.e ** (-3/( x**2)) ) )
    
for x in myfloatrange(minx, maxx+step, step):
    yseriesb.append( (0 if x == 0 else np.e ** (-2/( x**2)) ) )

for x in myfloatrange(minx, maxx+step, step):
    yseriesc.append( (0 if x == 0 else np.e ** (-1/( x**2)) ) )

xseries = [ x for x in myfloatrange(minx, maxx+step, step) ]

xseries

# create the plots
a, = plt.plot(xseries, yseriesa)
b, = plt.plot(xseries, yseriesb)
c, = plt.plot(xseries, yseriesc)

# using the colors of the plots (created by matplotlib),
# choose the function value at [0] to draw a horizonta line 
# from minx to maxx
plt.hlines( yseriesa[0], minx, maxx, colors = a.get_color() )
plt.hlines( yseriesb[0], minx, maxx, colors = b.get_color() )
plt.hlines( yseriesc[0], minx, maxx, colors = c.get_color() )

plt.show()
```
