```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np

data = {
    'A': [10,20,15],
    'B': [14,12,11],
    'C': [5,24,9]
}

labels = 'A', 'B', 'C'
sums = [ sum( data['A'] ), sum( data['B'] ), sum( data['C'] ) ]
explode = (0.1, 0.1, 0.1)

plt.pie(sums, explode=explode, labels=labels, autopct='%.0f%%',
        wedgeprops={'width':.8, 'edgecolor':'b'})

plt.axis('equal')

plt.show()
```
