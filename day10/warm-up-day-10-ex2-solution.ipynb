```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
'personid': [0,1,2,3,4,5],
'number of children': [2,3,1,2,1,3],
'number of adults': [2,2,1,2,2,1],
'household income': ['not specified', 40200, 120300, 'not specified', 55000, 100000]
})
```

```python
df_replaced = df.replace('not specified', np.nan)
df_replaced
```

```python
df_replaced['household income'].mean()
```

```python
df_fill = df_replaced.fillna(57617)
df_fill
```

```python
df_fill['household income'].mean()
```

```python
df_fill['income per person'] = df_fill['household income'] / ( df_fill['number of children'] + df_fill['number of adults'] )
df_fill
```

```python
df_fill.sort_values('income per person')
```
