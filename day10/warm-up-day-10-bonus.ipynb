```python
import pandas as pd

d = {
    'x': [100,10,200,300,100,100,500,600],
    'y': [10,100,300,200,100,100,10,600]
}

d = pd.DataFrame(d)
d
```

```python
samenumbers = d[d['x'] == d['y']]

# designate the firstset of variables to be the unique case where x = y
firstset = samenumbers[ samenumbers.duplicated() ]
```

```python
# get the indices of these rows where x = y
samenumbersindices = list(d[d['x'] == d['y']].index.values)
```

```python
# drop the x = y rows from the original dataframe
d = d.drop( samenumbersindices )
d
```

```python
# merge d on itself, left_on x and right_on y so that we can have possible candidates for symmetry on one row
firstmerge = d.merge(d, how='inner', left_on='x', right_on='y')
```

```python
firstmerge
```

```python
# determine if the candidate in column 'x_y' and 'y_y' is in fact a symmetrical pair
secondset = firstmerge[ ( firstmerge['x_x'] == firstmerge['y_y'] ) & ( firstmerge['y_x'] == firstmerge['x_y'] ) ]

# delete the columns we no longer need
del secondset['x_y']
del secondset['y_y']

# drop the duplicates
secondset = secondset.drop_duplicates()

# rename the columns for appending to the first set
secondset = secondset.rename(columns={'x_x':'x', 'y_x': 'y'})

secondset
```

```python
finalset = firstset.append(secondset, sort=True)

finalset
```

```python
finalset = finalset[finalset['x'] <= finalset['y']]

finalset
```
